# Decision Tree Regressor

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import accuracy_score, classification_report
from sklearn.datasets import load_iris
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt
from sklearn import tree
from sklearn.metrics import mean_squared_error,r2_score
from sklearn.linear_model import LinearRegression
```

```python
X,y = make_regression(n_samples=100,n_features=1,noise=0.1, random_state=42)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=13)
```

```python
dt = DecisionTreeRegressor(criterion='squared_error',splitter='best',max_depth=4)
lr = LinearRegression()
```

```python
dt.fit(X_train,y_train)
lr.fit(X_train,y_train)
```

```python
y_pred1 = dt.predict(X_test)
```

```python
y_pred2 = lr.predict(X_test)
```

```python
print("r2_Score Linear Regression: ", r2_score(y_test,y_pred1))
print("r2_Score Decision Tree: ", r2_score(y_test,y_pred2))
```

```python
plt.figure(figsize=(12,8))  
tree.plot_tree(dt, filled=True)
plt.show()
```

```python
plt.scatter(X, y, color="black", label="Data", alpha=0.5)
plt.plot(X_test, y_pred1, color="green", label="Decision Tree Prediction")
plt.plot(X_test, y_pred2, color="red", label="Linear Regression Prediction")

plt.title("Decision Tree vs Linear Regression")
plt.xlabel("Feature")
plt.ylabel("Target")
plt.legend()

# Show plot
plt.show()
```
